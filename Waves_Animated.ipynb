```python
from math import *
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

This notebook animates the function

$$
y = y_0 \cos(k x - \omega t)
$$

Choose your values for amplitude, wave number, and angular frequency here:

```python
k = 1.57
omega = 3.14
y0 = 1.0
```

```python
fig, ax = plt.subplots()

x = np.arange(0, 10, 0.1)
line, = ax.plot(x,y0*np.cos(k*x),'r.')
ax.set_xlabel('x (m)')
ax.set_ylabel('y (m)')
ax.grid(visible=True)


def animate(i):
    t=i/10
    line.set_ydata(y0*np.cos(k*x - omega*t))
    ax.set_title('t = '+str(t)+' s')
    return line,

anim = animation.FuncAnimation(fig,animate,interval=200,blit=True,save_count=200)

plt.close()
```
